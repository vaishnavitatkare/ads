```python
# Calculate the Interquartile Range (IQR) for 'total_bill' and 'tips'
iqr_total_bill = stats.iqr(df['total_bill'])
iqr_tips = stats.iqr(df['tip'])

print("IQR For Total Bill:", iqr_total_bill)  # IQR for 'total_bill'
print("IQR For Tips:", iqr_tips)  # IQR for 'tips'
```
